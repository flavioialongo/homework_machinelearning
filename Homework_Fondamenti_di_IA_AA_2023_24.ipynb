##Data Preparation


Funzioni necessarie per trasformare le variabili di tipo stringa in variabili numerali.

Per gli attributi MTRANS e NObesity sono presenti due funzioni che eseguono la One Hot Encoding. Per la regressione la One Hot Encoding viene effettiata sia per MTRANS che per NObesity; per la classificazione viene usata solo la One Hot Encoding in riferimento a MTRANS, l'attributo NObesity viene processato diversamente.

```python
#Funzioni ausiliarie per trasformare variabili categoriche
#in variabili numerali

features_to_transform=["Gender", "family_history_with_overweight", "FAVC", "SMOKE", "SCC", "CAEC", "CALC"]


def one_hot_transportation(dataset):
  dataset["mtr"]=dataset["MTRANS"]
  dataset = pd.get_dummies(dataset, columns=["MTRANS"], dtype=int)
  dataset = dataset.rename(columns={"mtr":"MTRANS"})
  return dataset

def one_hot_obesity(dataset):
  dataset["ob"]=dataset["NObesity"]
  dataset = pd.get_dummies(dataset, columns=["NObesity"], dtype=int)
  dataset = dataset.rename(columns={"ob":"NObesity"})
  return dataset

def binary_variable_to_integer(x):
  return 1 if (x=="yes" or x=="Male")  else 0

def categorical_value_to_int(x):
    if(x=="Sometimes"):
      return 1
    elif(x=="Frequently"):
      return 2
    elif(x=="Always"):
      return 3
    else:
      return 0

def dataset_to_numerical(dataset, attributes):
  ret = dataset.copy()
  for i in attributes:
    if(i == "CAEC" or i=="CALC"):
      ret[i] = ret[i].apply(categorical_value_to_int)
    else:
      ret[i] = ret[i].apply(binary_variable_to_integer)
  return ret

# 0 <= sovrappeso
# 1 > sovrappeso

def classification_nobesity(value):
    if(value== 'Insufficient_Weight' or value=='Normal_Weight' or value=="Overweight_Level_I" or value=="Overweight_Level_II"):
      return 0
    else:
      return 1

```

```python

# MinMax normalization
# Porta tutte le variabili nel range [0,1]
def normalize(dataset, exclude=[]):
  normalized = pd.DataFrame()
  for col in dataset.columns:
    if(col in exclude):
      normalized[col] = dataset[col]
      continue
    max = dataset[col].max()
    min = dataset[col].min()
    if((max-min)==0):
      max = 1
    normalized[col] = (dataset[col]-min)/(max-min)

  return normalized

# Funzione usata per denormalizzare i valori predetti
def denormalize(array, min, max):
  res = np.array([array]).reshape(-1)
  for i in range(len(array)):
    res[i]= array[i]*(max-min)+min
  return res
```

```python
# Copiamo il dataset originale, così da poter fare tutte le modifiche necessarie
work_dataset = dataset.copy()


# One Hot Encoding di MTRANS per l'intero dataset
work_dataset = one_hot_transportation(work_dataset)


# Trasformazione variabili di tipo stringa -> variabili numerali
regression_dataset = dataset_to_numerical(work_dataset.copy(), features_to_transform)


# One Hot Encoding NObesity per regressione
regression_dataset = one_hot_obesity(regression_dataset)

# Trasformazione variabili di tipo stringa -> variabili numerali
classification_dataset = dataset_to_numerical(work_dataset.copy(), features_to_transform)

# Trasformazione dei valori di NObesity in
# 0 <= sovrappeso
# 1 > sovrappeso
classification_dataset['NObesity'] = classification_dataset['NObesity'].apply(classification_nobesity)


```

##Linear Regressor

```python
from sklearn.metrics import mean_squared_error
class LinearRegressor:

  def __init__(self, epoch=0, step=0.01, minibatch=0, l2 = 0):
    self.epoch=epoch
    self.step=step
    self.minibatch=minibatch
    self.coefficient_value=[]
    self.l2=0
    self.fitted=False

  def _gradientDescent(self, x,y_true,epoch,step, minibatch, l2, printMse):
    x_number = x.shape[0]

    #inizializzo randomicamente i pesi
    weights = np.random.randn(x.shape[1])
    i=0

    # salvo la trasposta della matrice in input
    # così da non doverla caricare ogni volta (solo nel caso in cui non si usi un minibatch)
    x_transpose = x.T

    while(i<epoch):

      if(printMse and i%(epoch/10)==0):
        print("Epoch: "+str(i)+" MSE: "+str(mean_squared_error( denormalize(y_true, r_train_min, r_train_max), denormalize(np.dot(x, weights), r_train_min, r_train_max))))

      # se minibatch!=0 si esegue il gradient descent su un minibatch randomico del dataset
      if(minibatch!=0):
        permutation = np.random.permutation(x_number)
        X_shuffled = x[permutation]
        y_shuffled = y_true[permutation]

        for j in range(0, x_number, minibatch):
          end = j + minibatch

          # minibatch di x, y
          x_batch = X_shuffled[i:end]
          y_batch = y_shuffled[i:end]

          # il gradiente è pari a (-2/minibatch_size)*(X.t( Y - pred))
          error = y_batch-np.dot(x_batch, weights)
          gradient = (-2/minibatch)*np.dot(x_batch.T, error)

          #formula di aggiornamento pesi
          weights = weights - step*(gradient + l2*weights)

      else:
        # gradient descent su intero dataset
        error = y_true-np.dot(x, weights)
        gradient = (-2/x_number)*np.dot(x_transpose, error)
        weights = weights - step*( gradient + l2*weights)
      i+=1
    return weights



  def fit(self, x, y_true, printMse=True):

    # si inserisce una colonna dummy fatta di soli 1
    # che rappresenta il bias
    if 'Bias' not in x.columns:
      x.insert(0, column="Bias", value=1)
    x_t = np.array(x)
    y = np.array(y_true)
    # Se non sono specificate le epoche, si usa la Normal Equation
    if(self.epoch == 0):
      self.coefficient_value = np.dot(np.linalg.pinv(x_t), y);
    else:
      self.coefficient_value =  self._gradientDescent(x_t,y, self.epoch, self.step, self.minibatch, self.l2, printMse)
    self.fitted=True


  def predict(self, x):
    if not self.fitted:
      raise Exception("Fit the model first")
    if 'Bias' not in x.columns:
      x.insert(0, column="Bias", value=1)


    # Pred = X*W
    res = np.dot(x, self.coefficient_value)
    return denormalize(res, r_train_min, r_train_max)

```

##Logistic Classifier


```python
from sklearn.metrics import accuracy_score


class LogisticClassifier:

  def __init__(self, epoch=10000, step=0.01, l2=0, threshold=0.5):
    self.epoch = epoch
    self.step = step
    self.threshold = threshold
    self.coefficient_value = []
    self.l2 = 0
    self.fitted = False

  def _sigmoid(self, x):
    return 1 / (1 + np.exp(-x))


  def fit(self, x, y_true, printAccuracy=True):

    # Colonna dummy di soli 1 che rappresenta il bias
    if 'Bias' not in x.columns:
      x.insert(0, column="Bias", value=1)
    x_transpose = x.T
    weights = np.random.randn(x.shape[1])

    for i in range(self.epoch):
      if(printAccuracy and i%(self.epoch/10)==0):

        # predizione lineare
        y_hat = np.dot(x, weights)

        # viene passata nella sigmoide
        y_predicted = self._sigmoid(y_hat)

        #se y_predicted è maggiore della threshold allora è 1, altrimenti 0
        result = [1 if i > self.threshold else 0 for i in y_predicted]

        print("Epoch: "+str(i)+" Accuracy: "+str(accuracy_score(result, y_true)))

      # predizione lineare = X*W
      linear_prediction = np.dot(x, weights)

      # si passa la predizione lineare nella sigmoide
      prediction = self._sigmoid(linear_prediction)

      error = y_true - prediction

      # gradiente = -2/dataset_size*(X.t*(Y-y_pred))
      gradient = (-2 / x.shape[0]) * np.dot(x_transpose, error)

      #aggiornamento pesi
      weights -= self.step *(gradient) + self.step*(self.l2*weights)

    self.fitted = True
    self.coefficient_value=weights

  def predict(self, x):
    if(not self.fitted):
      raise Exception("Fit the model")
    if 'Bias' not in x.columns:
      x.insert(0, column="Bias", value=1)
    y_hat = np.dot(x, self.coefficient_value)
    y_predicted = self._sigmoid(y_hat)

    # 1 se la predizione supera la threshold
    # 0 altrimenti
    result = [1 if i > self.threshold else 0 for i in y_predicted]

    return np.array(result)

```

##K-Nearest-Neighbour Classifier

```python
from collections import Counter
class KNNClassifier:

  def __init__(self, k=3, distance_type=2):
    self.k = k
    self.distance_type = distance_type
    self.fitted = False


  # calcola distanza l2
  def _euclidean_distance(self,x1,x2):
    return np.linalg.norm(x1-x2)

  #calcola distanza l1
  def _manhattan_distance(self, x1,x2):
    return np.linalg.norm(x1-x2, 1)


  # prende in input una riga di un dataset
  # e ne calcola la distanza da tutte le righe
  # del dataset dataset di train
  def _compute_nearest(self, x):
    distance = []
    for x_train in self.X_train:
      if(self.distance_type == 2):
        distance.append(self._euclidean_distance(x, x_train))
      else:
        distance.append(self._manhattan_distance(x, x_train))

    # Ordina le distanze, da minore a maggiore e salva i corrispettivi indici
    # tra questi vado a prendere i primi k
    k_nearest_indices = np.argsort(distance)[:self.k]
    k_nearest_elements = self.Y_train[k_nearest_indices]


    # Prendo il valore con più occorrenze tra i k e lo fornisco in output
    most_common = Counter(k_nearest_elements).most_common(1)[0][0]

    return most_common


  def fit(self, x, y_true):
    if("Bias" in x.columns):
      self.X_train = x.drop(columns=["Bias"]).to_numpy()
    else:
      self.X_train = x.to_numpy()

    self.Y_train = y_true.to_numpy()

    if((self.k)%2==0):
      raise Exception("Insert an odd number for k")
    if(self.distance_type != 1 and self.distance_type!=2):
      raise Exception("Insert a valid distance metric (1 - 2)")
    self.fitted = True

  def predict(self, X):
    if("Bias" in X.columns):
      X_t = X.drop(columns=["Bias"]).to_numpy()
    else:
      X_t = X.to_numpy()
    if(not self.fitted):
      raise Exception("Fit the model first")

    # passo di volta in volta le righe dell'input X
    # alla funzione compute_nearest

    predict = [self._compute_nearest(x) for x in X_t]
    return np.array(predict)
```

```python
import math


class DecisionTree:


  # Classe per rappresentare i vari nodi di un albero
  class TreeBranch:
    def __init__(self, split_attribute_value, subtree):
      self.subtree = subtree
      self.split_attribute_value = split_attribute_value

  # Classe che modella i branch di un albero
  # ogni Branch ha un attributo di split (es. Weight)
  # un sottoalbero se non è una foglia e un valore (output) se è una foglia
  class DTreeStructure:
      def __init__(self, split_attribute="", *, value=None):
        self.split_attribute = split_attribute
        self.value = value
        self.branches=[]

      def add_branch(self, branch):
        self.branches.append(branch)

      def is_leaf_node(self):
        return self.value is not None

  def __init__(self,max_depth=10000, target="NObesity"):
    self.max_depth = max_depth
    self.target = target
    self.tree = []


  def fit(self, X):
    self.X_train = X
    self.n_features = X.shape[1]
    self.n_examples = X.shape[0]

    # per evitare che nella lista di attributi su cui fare un possibile split
    # sia presente l'attributo target
    if(self.target in X.columns):
      self.attributes = set(X.columns.drop(self.target))
    else:
      self.attributes = set(X.columns)

    # funzione ricorsiva che crea il vero Decision Tree
    self.tree = self._grow_tree(X, self.attributes, None)



  def _grow_tree(self, examples, attributes, parent_examples, depth=0):

    # Non ci sono esempi con questa combinazione di attributi -> si prende il valore più comune dell'albero padre
    if examples.empty:
      return self._most_common_label(parent_examples)

    # Non ci sono più attributi su cui splittare oppure abbiamo raggiunto la profondità massima ->  si prende il valore più comune
    if(not attributes or (depth>=self.max_depth)):
      return self._most_common_label(examples)

    # Si hanno soltanto esempi con la stessa classificazione -> si prende il valore più comune
    if(len(examples[self.target].unique())==1):
      v=examples[self.target].iloc[0]
      return self.DTreeStructure(value=v)

    # si prende l'attributo che ha più importanza
    split_attribute = self._importance(examples, attributes)

    # non si è trovato l'attributo, non si può splittare -> si prende il valore più comune
    if split_attribute=='':
      return self._most_common_label(examples)


    #
    tree = self.DTreeStructure(split_attribute)

    split_attribute_values = examples[split_attribute].unique()

    for value in split_attribute_values:

      # Si prendono solo gli esempi che hanno lo stesso value per l'attributo split_attribute
      exs = examples.loc[examples[split_attribute] == value]

      # Si rimuove l'attributo scelto dal set degli attributi
      attr = attributes - {split_attribute}

      # Generare il sottoalbero ricorsivamente
      subtree = self._grow_tree(exs, attr, examples, depth+1)

      # Aggiungere il sottoalbero nella variabile branch del TreeBranch
      tree.add_branch(self.TreeBranch(value, subtree))

    return tree

  def predict(self, X):
    ret = []

    # Per ogni riga di X calcola la predizione
    for index,row in X.iterrows():
      ret.append(self._predict_tree(row,self.tree))
    return ret


  def _predict_tree(self,sample,tree):

    index = 0

    # Si è arrivati alla foglia, viene returnata
    if(tree.is_leaf_node()):
      return tree.value


    # Ciclo su tutti i sottoalberi di tree
    for i in range(len(tree.branches)):
      # Se il valore del sample per l'attributo split_attribute è uguale a split_attribute_value
      # abbiamo trovato il sottoalbero giusto, si salva quindi l'indice
      if tree.branches[i].split_attribute_value==sample.loc[tree.split_attribute]:
        index = i
        break

    # si prende il sottoalbero corrispondente all'indice trovato prima
    new_tree = tree.branches[index].subtree

    # chiamata ricorsiva
    return self._predict_tree(sample, new_tree)


  # Information gain = Entropy - Remainder
  def _information_gain(self, X, attribute):
    information_gain = self._entropy(X)-self._remainder(X, attribute)
    return information_gain

  def _remainder(self,X,attribute):

    remainder=0


    # si prendono gli esempi del dataset con l'attributo attribute
    groups=X.groupby(attribute)


    for value,group in groups:

      # per evitare divisione per 0
      if(group.shape[0]!=0):

        # Remainder = (dim_gruppo/dim_totale_dataset)*entropia(gruppo)
        remainder+=(group.shape[0]/X.shape[0])*self._entropy(group)

    return remainder


  def _entropy(self, X):

    # No entropia, informazione precisa
    if (len(X[self.target].unique())==1):
      return 0

    # i valori sono solo 0,1 -> si usa la somma per trovare il numero di valori positivi
    sum=X[self.target].sum()


    probability=sum/(X.shape[0])

    # formula entropia
    entropy = -probability*math.log2(probability)-(1-probability)*math.log2(1-probability)
    return entropy


  def _importance(self, X, attributes):

    max = 0
    split_attribute = ""


    # Ciclo su ogni attributo e calcolo quale di questi fornisce il maggior gain
    # mediante funzione information_gain
    for attr in attributes:
      gain = self._information_gain(X, attr)
      if(gain>max):
        max = gain
        split_attribute = attr
    return split_attribute


  def _most_common_label(self, X):

    # Ritorna una foglia con il valore uguale al più comune in X
    counter = Counter(X[self.target])
    ret = counter.most_common(1)[0][0]
    return self.DTreeStructure(value=ret)





```

## Neural Network



Nell'inizializzazione dei pesi vengono usate formule non viste a lezione (Inizializzazione He e inizializzazione Xavier). Questo perché effettuando alcuni test, generando valori totalmente randomici per i pesi, si ottenevano più frequentemente NaN in output, anche con piccoli valori di learning rate.

```python
class NeuralNetwork:

  class Layer:

    def __init__(self, activation, input_size, output_size):



      if(activation=="relu"):
        stddev = np.sqrt(2. / input_size)  # He initialization
        self.weights = np.random.randn(input_size, output_size)*stddev
      else:
        stddev = np.sqrt(2. / (input_size + output_size)) # Xavier initialization
        self.weights = np.random.randn(input_size, output_size)*stddev
      self.bias = np.zeros(output_size)


  def __init__(self, layers, activation="relu", epoch=10000, alpha=0.01, l2=0.01, minibatch=0):

    if(type(layers) is not list ):
      raise Exception("Layer should be an array to indicate the structure of the Neural Network")


    if(activation=="relu"):
      self.activation = self._relu
    elif(activation=="sigmoid"):
      self.activation = self._sigmoid
    else:
      raise Exception("Insert valid activation function [relu, sigmoid]")

    self.epoch=epoch
    self.alpha=alpha
    self.l2=l2
    self.minibatch=minibatch


    self.network = []
    for i in range(len(layers)-1):
        self.network.append(self.Layer(activation, layers[i], layers[i+1]))

  def fit(self, X, y, print_mse=True):

    # viene rimossa la colonna dummy bias
    # perchè i bias vengono calcolati volta per volta nei layer

    if("Bias" in X.columns):
      X_without_bias = X.drop(columns=["Bias"])
      X_t = np.array(X_without_bias)
    else:
      X_t = np.array(X)
    y_t = np.array(y).reshape(len(y), 1)
    x_number = X_t.shape[0]
    predictions = np.random.randn(y_t.shape[0],1).reshape(-1)


    for i in range(self.epoch):

      # se minibatch = 0 il gradient descent viene effettuato su un minibatch del dataset iniziale
      if(self.minibatch!=0):
        if(print_mse and i%(self.epoch/10)==0):
            print("Epoch: "+str(i))

        # creazione dataset randomico di dimensione = minibatch
        permutation = np.random.permutation(x_number)
        X_shuffled = X_t[permutation]
        y_shuffled = y_t[permutation]

        for j in range(0, x_number, self.minibatch):
          end = j + self.minibatch
          x_batch = X_shuffled[j:end]
          y_batch = y_shuffled[j:end]

          # #step di feedforward e backpropagation per trovare i migliori pesi e bias
          outputs=self._feedforward(x_batch)
          self._backpropagation(self.network, outputs, y_batch, self.minibatch)

      else:
          if(print_mse and i%(self.epoch/10)==0):
            mse = mean_squared_error(denormalize(y_t.reshape(-1), r_train_min, r_train_max), denormalize(predictions, r_train_min, r_train_max))
            print("Epoch: "+str(i)+" ---- MSE: "+ str(mse))

          #step di feedforward e backpropagation per trovare i migliori pesi e bias
          outputs=self._feedforward(X_t)
          predictions = outputs[-1].reshape(-1)
          self._backpropagation(self.network, outputs, y_t, X_t.shape[0])
      self.fitted = True

  def predict(self, X):
    if("Bias" in X.columns):
      X_t = np.array(X.drop(columns=["Bias"]))
    else:
      X_t = np.array(X)
    if(not self.fitted):
      raise Exception("Fit the model first")
    return denormalize(self._feedforward(X_t)[-1].reshape(-1), r_train_min, r_train_max)


  def _relu(self, x, derivative=False):

    if(derivative):
      return np.where(x > 0, 1, 0)

    return np.maximum(0, x)


  def _sigmoid(self, x, derivative=False):
    if(derivative):
      sig = self._sigmoid(x, False)
      return sig*(1-sig)
    return 1 / (1 + np.exp(-x))

  def _feedforward(self, input):


    # Per ogni layer (eccetto l'ultimo) viene calcolato f = X*W + b
    # Viene passata f nella non linearità (sigmoide o relu)
    # e fornita in input al prossimo layer

    outputs = list([np.array(input)])
    for i in range(len(self.network)-1):
      weights = self.network[i].weights
      bias = self.network[i].bias
      output = self.activation(np.dot(outputs[i],weights)+bias)
      outputs.append(output)



    # L'ultimo layer calcola f = X*W + b ma non viene passata nella non linearità
    output=np.dot(outputs[-1],self.network[-1].weights)+self.network[-1].bias
    outputs.append(output)

    return outputs


  def _backpropagation(self, network,outputs,y, batch_size):


    # Per l'algoritmo di backpropagation vengono calcolati i vari gradienti

    # Sfrutta la chain rule:


    # Si definiscono
    # Theta = sigmoid / relu
    # f_i = h_i * W_i + b_i
    # h_0 = input, h_i = Theta(f_i)
    #
    # Si prende una variabile ausiliaria delta = dL/dfi
    # delta = -2(y-pred) se i è un layer esterno, altrimenti delta = Theta'(fi)*dL/df_i+1 dot (W_i+1.T)

    # dL/dWi = h_i.T dot delta
    # dL/dBi = delta dot 1 dove 1 è il vettore unitario, quindi dL/dBi = somma righe di delta

    for i in reversed(range(len(self.network))):
      activated=outputs[i+1]
      h_i=outputs[i]
      if(i==(len(self.network)-1)):
        delta = -(y-activated)
      else:
        activation_derivative = self.activation(activated, True)
        delta = activation_derivative * np.dot(delta, self.network[i+1].weights.T)

      gradient_weights =  (1/batch_size)*np.dot(h_i.T, delta)
      gradient_biases = (1/batch_size)*np.sum(delta, axis=0)


      network[i].bias = network[i].bias - self.alpha * gradient_biases
      network[i].weights = network[i].weights - self.alpha*(gradient_weights + 2*self.l2 * network[i].weights)


```

##Hyperparameter Tuning

Viene utilizzata una GridSearch che, presi in input una serie di iperparametri, ne calcola tutte le possibili combinazioni e allena il modello desiderato con ciascuna di queste combinazioni. Per i regressori sceglie gli iperparametri che forniscono il MSE più basso, mentre per i classificatori vengono scelti gli iperparametri che portano l'accuratezza maggiore. <br>

Per la regressione, alcune combinazioni di parametri potrebbero far divergere il gradiente e quindi darebbero luogo a NaN in output. Per aggirare questo problema, viene usata una try-except che salta ai prossimi parametri se per quelli attuali è presente un errore.

```python
import warnings
## per ignorare i warning di numpy quando il gradiente diverge
warnings.filterwarnings('ignore')

import itertools

def grid_search(model_type, X_train=[], y_train=[], X_test=[], y_test=[], params={}):
    best_score = None
    chosen_params = None
    classification=False
    # generare tutte le possibili combinazioni di iperparametri dal dizionario
    h_params_comb = list(itertools.product(*params.values()))
    for param_combination in h_params_comb:

        # printa i parametri attuali
        for i,key in enumerate(params.keys()):
          print(key,': ', param_combination[i],end='; ',sep='')

        # instanzia il modello cercato

        # è tutto racchiuso in una try-except perché
        # per alcuni parametri il gradiente potrebbe divergere e dar luogo a NaN in output
        # quindi per evitare l'interruzione dell'esecuzione quando viene fuori un errore
        # passo ai prossimi parametri
        try:
          if(model_type == "linear_regressor"):
            model = LinearRegressor(*param_combination)
            model.fit(X_train, y_train, False)
          elif(model_type == "logistic_classifier"):
            model = LogisticClassifier(*param_combination)
            model.fit(X_train, y_train, False)
            classification = True
          elif(model_type == "knn_classifier"):
            model = KNNClassifier(*param_combination)
            model.fit(X_train, y_train)
            classification = True
          elif(model_type == "decision_tree"):
            model = DecisionTree(*param_combination)
            model.fit(X_train)
            classification = True
          elif(model_type == "neural_network"):
            model = NeuralNetwork(*param_combination)
            model.fit(X_train, y_train, False)
          else:
            raise Exception("Unknown model type")
            return
          prediction = model.predict(X_test)
          # stampare i parametri per questa iterazione


          # printa lo score in base al tipo di modello (Classificatore/Regressore)
          if(classification):
            print('Accuracy:',end=' ')
            score=accuracy_score(y_test,prediction)
            print(score)
            if best_score is None or score > best_score:
              best_score = score
              chosen_params = param_combination
          else:
            print('MSE:',end=' ')
            score=mean_squared_error(prediction,y_test)
            print(score)
            # aggiornare il miglior punteggio e i migliori parametri se il punteggio corrente è migliore
            if best_score is None or score < best_score:
              best_score = score
              chosen_params = param_combination

        except Exception as e:
          print(e)
          continue

    return chosen_params
```
